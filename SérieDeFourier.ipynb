```python
import numpy as np
import matplotlib.pyplot as plt

# Definir a funo S(t) da srie de Fourier
def S(t, n_max):
    result = 0
    for n in range(1, n_max+1):
        result += 20/np.pi * (np.sin((2*n-1)*np.pi*t) / (2*n - 1))
    return result+5

# Definir o intervalo de tempo
t = np.linspace(-5, 5, 1000)

# Calcular a srie de Fourier para diferentes valores de n_max
n_max_values = [1, 2, 5, 10, 20, 100, 1000]  # Valores mximos de n para calcular
fig, axs = plt.subplots(len(n_max_values), 1, figsize=(8, 6*len(n_max_values)))

# Calcular e plotar os grficos para cada valor de n_max
for i, n_max in enumerate(n_max_values):
    ax = axs[i]
    ax.plot(t, S(t, n_max))
    ax.set_title(f'Srie de Fourier para n_max = {n_max}')
    ax.set_xlabel('t')
    ax.set_ylabel('S(t)')

plt.tight_layout()
plt.show()
```
